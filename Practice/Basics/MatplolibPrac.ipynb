```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2,100) #generates 100 data between 0 and 2

plt.figure(figsize=(5, 2.7), layout='constrained') 
#plt.figure() -> initializes a window/ canvas
#figsize=(5,2.7) canvas size: (width, height)= (5,2.7)inch
#layout='constrained' : Automatically adjusts padding and spacing around the subplots so that the subplots don't overlap
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# Monthly data 
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] )
# Sales data (in thousands) 
product_a = np.array([23, 25, 28, 30, 35, 38, 42, 45, 48, 50, 53, 55]) 
product_b = np.array([30, 32, 31, 33, 35, 37, 39, 41, 43, 45, 47, 50]) 
product_c = np.array([15, 18, 22, 25, 28, 32, 35, 38, 42, 45, 48, 52])

fig, axs= plt.subplots(2,2, figsize=(8,8), layout='constrained')

#first plotting
axs[0,0].plot(months, product_a, color='blue', label='product_a')
axs[0,0].plot(months, product_b, color='green', linestyle='--', label='product_b')
axs[0,0].plot(months, product_c, color='red', linestyle='--', label='product_c')
axs[0,0].set_xlabel('Month')
axs[0,0].set_ylabel('Sales (in thousands)')
axs[0,0].set_title('Monthly Sales Trends 2024')
axs[0,0].grid(True)
axs[0,0].legend()

#second plotting
container= axs[0,1].bar(['A','B','C'],[product_a[11], product_b[11], product_c[11]],color=['blue','green','red'])
axs[0,1].set_ylim(0,max([product_a[11], product_b[11], product_c[11]])*1.1)
axs[0,1].set_xlabel("Product")
axs[0,1].set_ylabel("Sales (in thousands)")
axs[0,1].set_title("December Sales Comparison")
axs[0,1].grid(True, axis='y', alpha=0.5)
axs[0,1].bar_label(container)

#third plotting
axs[1,0].scatter(months, product_a, color='blue', s=50)
axs[1,0].set_title( "Product A Sales Distribution")
axs[1,0].set_xlabel("Month")
axs[1,0].set_ylabel("Sales (in thousands)")
axs[1,0].grid(True, alpha=0.2)

#fourth plotting
quarters_A= product_a.reshape(4,3).sum(axis=1)
quarters_B= product_b.reshape(4,3).sum(axis=1)
quarters_C= product_c.reshape(4,3).sum(axis=1)
quarters=['Q1','Q2','Q3','Q4']
axs[1,1].bar(quarters, quarters_A, color='blue', label='Product_A')
axs[1,1].bar(quarters, quarters_B, bottom=quarters_A, color='green', label='Product_B')
axs[1,1].bar(quarters, quarters_C, bottom=quarters_A+quarters_B, color='red', label='Product_C')
axs[1,1].set_title("Quarterly Sales (Stacked)")
axs[1,1].set_xlabel("Quarter")
axs[1,1].set_ylabel("Total Sales (in thousands)")
axs[1,1].legend()
axs[1,1].grid(True, axis='y', alpha=0.5)

plt.show()
```
