```python
class Employee:
    class_name='Employee' #class attribute-- like static attribute

    #double underscore method/ dunder method
    def __init__(self,name,age,position): #attributes created in __init__ are called instance attributes
        self.name=name
        self.age=age
        self.position=position
    def __str__(self):
        return f"Name: {self.name}\nAge: {self.age}\nPosition: {self.position}"
    def __repr__(self):
        return "I am a representation."
    
emp=Employee("A",10,"B")
#the purpose of __str__ and __repr__ are different
#__str__ is used for pretty output
#__repr__ is used for debugger purpose, minimal output
print(emp)#__str__ will be used
emp # __repr__ will be used
    
```

```python
# Operator overdrive
class Vector:
    def __init__(self,a=0,b=0):
        self.a=a
        self.b=b
    def __add__(self,other):
        return Vector(self.a+other.a, self.b+other.b)
    def __sub__(self, other):
        result= Vector()
        result.a=self.a-other.a
        result.b=self.b-other.b
        return result # __mul__ and __truediv__
    def __repr__(self):
        return f"{self.a}i + {self.b}j"
v1,v2=Vector(10,20),Vector(30,40)
print(v1+v2)
print(v1-v2)
```

Interpolation:
Puppose of interpolation:
given x=[...] and y=[...] the goal is to create a function y=f(x) so that one can plug in any value for y for any x


```python
import matplotlib.pyplot as plt
x=np.linspace(0,5,5)
y=x**2
plt.figure(figsize=(4,4),layout='constrained')
plt.scatter(x,y)
plt.show()
```

Suppose we want to get f(x=1.5). One of the best ways to get that is to use interpolation


One of the most basic form of interpolation is to just connect the data points using straight lines. Actually this is what plt.plot() does


```python
plt.figure(figsize=(4,4),layout='constrained')
plt.plot(x,y, linestyle='--', marker='o')
plt.show()
```

This is linear interpolation


```python
#we will find f(x=1.5) and f(x=-0.5) and f(x=6)
new_x=np.array([-0.5,1.5,6])
new_x=np.interp(new_x, x,y)
print(f"f(x=1.5)= {new_x[1]}")
print(f"f(x=-0.5)= {new_x[0]}")
print(f"f(x=6)= {new_x[2]}")
```
