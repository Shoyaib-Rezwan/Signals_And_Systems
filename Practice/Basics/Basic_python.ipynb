Sorting iterables (list, tuple, dictionary, str.....)


Inplace sorting


```python
a=[5,7,8,9,5,8,9,9,1,2,5]
a.sort()
print(a)
a=[5,7,8,9,5,8,9,9,1,2,5]
a.sort(reverse=True)
print(a)
```

Immutables like tuple, str, dict doesn't support inplace sortings :')


```python
a=(5,7,8,9,5,8,9,9,1,2,5)
a=sorted(a)
print(a)
a="jhbgfsfmSMIGuzhdg"
a=sorted(a, reverse=True)
print(a)
```

Sorting dicts


```python
a={'fngd':10, 'kjdngs':25, 'lsndgf sv':-69, 'aakmnd qq': 12}
print(sorted(a)) #it will only return the sorted keys
print(sorted(a.items())) # a.items() will return the dictonary items as a list of tuples... sorted based on first entries of the tuples i.e. the keys
print(dict(sorted(a.items()))) # a.items() will return the dictonary items as a list of tuples. So, you should typecast to dictinary 
b=dict(sorted(a.items(), key=lambda item:item[1] )) #based on the keys lambda==> short form of function declaration
print(b)
```
