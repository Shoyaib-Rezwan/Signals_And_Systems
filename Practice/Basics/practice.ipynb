```python
class Employee:
    class_name="Employee"
    def __init__(self,name,age, position):
        self.name=name
        self.age=age
        self.position=position
    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Position: {self.position}"
    def __repr__(self):
        return self.name+" "+str(self.age)+" "+str(self.position)
```

```python
emp=Employee("Shoyaib", 50, "Teacher")
print(emp)
emp
```

```python
class Complex:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self, other):
        return Complex(self.a+other.a, self.b+other.b)
    def __sub__(self, other):
        return Complex(self.a-other.a, self.b-other.b)
    def __repr__(self):
        return f"{self.a} + j{self.b}"
a=Complex(10,20)
b=Complex(20,-10)
print(a+b)
print(a-b)
```
